## Medals

```python
## a
medals['Rank'] = medals['Total'].rank(ascending=False)
medals.sort_values(by='Total', ascending=False)
```

```python
## b
top_gold = medals.nlargest(10, 'Gold')
top_silver = medals.nlargest(10, 'Silver')
top_bronze = medals.nlargest(10, 'Bronze')

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].bar(top_gold['NOC'], top_gold['Gold'], color='gold')
axes[0].set_title('Top 10 Countries by Gold Medals')
axes[0].set_xlabel('Country')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=90)

axes[1].bar(top_silver['NOC'], top_silver['Silver'], color='silver')
axes[1].set_title('Top 10 Countries by Silver Medals')
axes[1].set_xlabel('Country')
axes[1].set_ylabel('Count')
axes[1].tick_params(axis='x', rotation=90)

axes[2].bar(top_bronze['NOC'], top_bronze['Bronze'], color='#cd7f32')
axes[2].set_title('Top 10 Countries by Bronze Medals')
axes[2].set_xlabel('Country')
axes[2].set_ylabel('Count')
axes[2].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

## More than one data file

```python
## a
count = 0
for f in athletes['NOC'].unique():
    if f not in medals['NOC'].unique():
        count += 1
print(count)
```

```python
## b
signed_up_counts = athletes.groupby('Discipline').size().reset_index(name='Signed Up')
total_entrants = entries[['Discipline','Total']].rename(columns={'Total': 'Entries'})

merged = pd.merge(signed_up_counts, total_entrants)
merged['Diff'] = merged['Signed Up'] - merged['Entries']
merged.to_csv('discipline_signed_up_diff.csv')
merged.head()
```

```python
## c
total_medal_counts = pd.merge(athletes.groupby('NOC').size().reset_index(name='Athletes'),
                     medals[['NOC', 'Total']])

total_medal_counts['Ratio'] = total_medal_counts['Total'] / total_medal_counts['Athletes']
total_medal_counts = total_medal_counts.sort_values(by='Ratio', ascending=False)

total_medal_counts.head(10)
```
