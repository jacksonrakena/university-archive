#### (b)

```python
# Similarities:
# - Both arrays and lists can store numerical elements
# - Both arrays and lists store items in order
# Differences:
# - Lists can store elements of different data types, while numpy arrays can only store
# one type of element at a time
# - Because numpy arrays are optimised for integers, they are much faster than Python lists at performing
# mathematical operations
# - Numpy arrays have better support for two-dimensions, including printing them to the console in a 
# readable format

# Numpy arrays are more powerful, as they have basically the same feature set as python lists,
# but have a lot more powerful features important for data science, like cross-products, dot-products,
# easy handling of N-dimensions, multiplying, and they are also nice to look at when printed.

two_dimensional_array = np.random.random((3,3))
print("Easily print two dimensional array:")
print(two_dimensional_array)

print("Make another one:")
another_one = np.random.random((3,3))
print(another_one)
multiplied = another_one * two_dimensional_array
print("Multiply them together easily:")
print(multiplied)
```

#### (f)

```python
def find_sunspots(month, year):
    if month > 12 or month < 1 or year > 1983 or year < 1749:
        return 0
    observation = ((year-1749)*12)+(month-1)
    return sunspots[observation]

print(find_sunspots(1, 1970))
print(find_sunspots(4, 1876))
print()
for i in range(1,13):
    print(f"Month {i}, 1900: {find_sunspots(i, 1900)}")
print()
print(find_sunspots(1, 1749))
print(find_sunspots(12, 1983))
```

#### (g)

```python
def all_sunspots_in_year(year):
    return sunspots[((year-1749)*12):((year-1749)*12)+12]

print(all_sunspots_in_year(1795))
print(all_sunspots_in_year(1915))
print()
# Boundary cases
print(all_sunspots_in_year(1749))
print(all_sunspots_in_year(1983))
```

#### (d)

```python
largest_diff_n = -1
m = -1
for month in range(1, 13):
    s_1800 = find_sunspots(month, 1800)
    s_1900 = find_sunspots(month, 1900)
    diff = abs(s_1800-s_1900)
    if diff > largest_diff_n:
        largest_diff_n = diff
        m = month

print(f"Largest differential is {largest_diff_n} in month {m} - 1800: {find_sunspots(m, 1800)}, 1900: {find_sunspots(m,1900)}")
```

#### (f)

```python
import math
def get_year_and_month(count):
    for i, data in enumerate(sunspots):
        if data == count:
            year = (math.floor(i/12))+1749
            month = (i % 12)+1
            print(f"Count {count} matches month {month} of year {year}")

for i, data in enumerate(sunspots):
    if data > 200:
        get_year_and_month(data)
```

#### (g)

```python
def get_month_name(month_index):
    return ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"][month_index-1]
    
for year in range(1900, 1911):
    largest_month = -1
    for month in range(1, 13):
        if largest_month == -1 or find_sunspots(month, year) > find_sunspots(largest_month, year):
            largest_month = month
    print(f"Most sunspots in {year}: {get_month_name(largest_month)} with {find_sunspots(largest_month, year)} sunspots")
```

### 2.4

```python
import matplotlib.pyplot as plt
```

#### (a)

```python
plt.title("Sunspot counts")
plt.plot(sunspots, '--')
plt.ylabel('Number of sunspots measured')
plt.xlabel('Month (0 = January, 1749)')
```

#### (b)

```python
sunspots_1800_1850 = sunspots[((1800-1749)*12):((1850-1749)*12)+12]
sunspots_1900_1950 = sunspots[((1900-1749)*12):((1950-1749)*12)+12]

plt.title("Sunspot counts over time in two different periods")
plt.plot(sunspots_1800_1850, 'r--')
plt.plot(sunspots_1900_1950, 'g-.')
plt.ylabel('Number of sunspots measured')
plt.xlabel('Month in time span (see legend)')
plt.legend(['1800-1850','1900-1950'])
plt.savefig('2.4(b).png')
```

#### (c)

```python
fig, axs = plt.subplots(1,2)
axs[0].title.set_text('Sunspots, 1800-1850')
axs[1].title.set_text('Sunplots, 1900-1950')
axs[0].plot(sunspots_1800_1850, 'r--')
axs[1].plot(sunspots_1900_1950, 'g-.')

axs[0].set_xlabel('Month after 1800')
axs[1].set_xlabel('Month after 1900')
axs[0].set_ylabel('Sunspot count')
axs[1].set_ylabel('Sunspot count')
plt.savefig('2.4(c).png')
```

#### (d)

```python
sunspots_19s = sunspots[((1801-1749)*12):((1900-1749)*12)+12]
print(sunspots_19s[0:13])
sunspots_19s_shifted = np.roll(sunspots_19s, 12)
sunspots_19s_shifted[0:12] = 0
print(sunspots_19s_shifted[0:13])

plt.title("Sunspot counts in 19th century, with and without lag")
plt.plot(sunspots_19s, 'r--')
plt.plot(sunspots_19s_shifted, 'g-.')
plt.ylabel('Number of sunspots measured')
plt.xlabel('Month in time span (see legend)')
plt.legend(['No lag','12 month lag'])
```
