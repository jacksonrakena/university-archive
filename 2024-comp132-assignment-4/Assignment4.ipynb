### (h)

```python
def calc_crime_rate(row):
    if row['Population'] == 0 and row['Victimisations'] == 0:
        return 0
    if row['Population'] == 0 and row['Victimisations'] != 0:
        return np.nan
    return (row['Victimisations'] / row['Population']) * 100
crime_rate = v.apply(calc_crime_rate, axis=1)
v['CrimeRate'] = crime_rate
```

```python
wellington.head(5)
```

### (b, c)

```python
import matplotlib.pyplot as plt

urbans = []
pops = []
vics = []
for urban in wellington['Urban'].unique():
    urbans.append(urban)
    pops.append(wellington[wellington['Urban']==urban]['Population'].sum())
    vics.append(wellington[wellington['Urban']==urban]['Victimisations'].sum())

fig, ax1 = plt.subplots(figsize=(12,8))
plt.xlabel('Urban zone')
plt.xticks(rotation='vertical')
width = 0.25
br1 = np.arange(len(urbans))
br2 = [x + width for x in br1]

v = ax1.bar(br1, vics, tick_label=urbans, color='r', width=width, label='Victimisations')
ax1.set_ylabel('Victimisations (total)')
plt.xticks(rotation='vertical')

ax2 = ax1.twinx()
ax2.set_ylabel('Population')
p = ax2.bar(br2, pops, tick_label=urbans, color='b', width=width, label='Population')


plt.legend(handles=[v,p])
plt.title('Victimisations and population counts in various urban areas of Wellington Region')

plt.show()
```
